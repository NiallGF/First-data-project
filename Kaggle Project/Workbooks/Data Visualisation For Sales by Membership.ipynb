```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/workspaces/Practise-Code/Kaggle Project/Data/supermarket_sales_clean.csv')
```

```python
df_sales_by_memebership = df.groupby('Customer type')['Total'].sum().reset_index()
df_sales_by_memebership.plot(kind='bar', x='Customer type', y='Total', color=['skyblue', 'red'])
```

```python
df_mean_customer_spend = df.groupby('Customer type')['Total'].mean().reset_index()
df_mean_customer_spend.plot(kind='bar', x='Customer type', y='Total', color=['skyblue', 'red'])
```
