```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/workspaces/First-data-project/Kaggle Project/Data/Cleaned Data/df_cleaned.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_price_by_bedrooms = df.groupby('bedrooms')['price'].mean().reset_index() 

plt.figure(figsize=(12, 6))

sns.barplot(data=df_price_by_bedrooms, x='bedrooms', y='price', palette='viridis')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Pricein $')
plt.title('Average Price by Number of Bedrooms')
```

**Bedrooms Effect on Mean Price**

- The analysis still shows a very positive correlation between the number of bedrooms and mean price.
- However, the price gaps between certain bedroom counts vary:
    - The gaps between one and two bedrooms, and between three and four bedrooms, are much larger than between two and three bedrooms, or between four and five bedrooms.
    - This suggests the market places more value on two-bedroom properties over one-bedroom homes, and four-bedroom homes over three-bedroom ones, but less so in the transitions between two and three bedrooms, or four and five bedrooms.
- Other factors could also influence this trend, and further analysis on those factors is forthcoming.
- The correlation between bedrooms and price appears less strong compared to bathrooms, as adding more bedrooms does not consistently result in the same price jump per bedroom. This explains why the correlation with bedrooms is weaker than with bathrooms, where price increments are more even between each additional bathroom.


```python
df_bedrooms_to_squarefeet = df.groupby('bedrooms')['sqft_living'].mean().reset_index()

plt.figure(figsize=(12, 6))

sns.barplot(data=df_bedrooms_to_squarefeet, x='bedrooms', y='sqft_living', hue='bedrooms', palette='viridis') 
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Square Feet')
plt.title('Average Square Feet by Number of Bedrooms')
```

**Square Footage (Living Area) to Bedrooms**

- The small price difference between four- and five-bedroom properties could be explained by the relatively small difference in square footage between these two categories, as shown in the graph above.
- The gap in living space between one- and two-bedroom homes is also smaller than one might expect, potentially contributing to the smaller price difference between these two categories.
- Overall, there isn’t much more to highlight, as the differences in size between bedroom counts tend to align with the observed pricing patterns.
