```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/workspaces/Practise-Code/Kaggle Project/Data/Cleaned Data/df_cleaned.csv')

df.rename(columns={'date': 'date'}, inplace=True)

df['date'] = pd.to_datetime(df['date'])
```

<h1>Boxplots and individual anlysis<h1>

```python
# Boxplot for 'price'
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['price'])
plt.title('Boxplot of Price')
plt.ylabel('Price in Million $')
plt.show()
```

### Price

- **Median:** The median is closer to the lower quartile (Q1), indicating that a significant portion of the houses are priced near the lower end of the market.

- **Box Size:** The small size of the box indicates that a large number of houses have prices close to the median, showing low variability in this central range of prices.

- **Skewness:** The distribution is right-skewed, evidenced by a long upper whisker. This suggests that while most properties are priced within a narrow, affordable range, there are a few properties with much higher prices.

- **Outliers:** Above the upper whisker, there is a small stretch of outliers, representing properties with significantly higher prices, likely luxury or high-demand homes.

- **Interpretation:** The right skewness with a long top whisker and concentrated box suggests that the majority of the market is relatively affordable, but there exists a distinct segment of higher-priced properties that push the distribution to the right.



```python
# Boxplot for 'sqft_living'
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['sqft_living'])
plt.title('Boxplot of Sqft Living')
plt.ylabel('Sqft Living')
plt.show()
```

### Sqft Living

- **Median:** The median square footage of living space is centrally located within the box, suggesting that the distribution of home sizes is fairly balanced around the median.

- **Box Size:** The box is not large, indicating that many properties have living spaces close to the median value.

- **Skewness:** The distribution shows a slight right skew, as indicated by the top whisker being slightly longer than the lower whisker. However, the lower whisker is only slightly shorter, indicating some stretching of the lower quartile as well. This suggests that while most homes are near the median size, there is some spread both below and above the median, with a slight emphasis on larger houses.

- **Outliers:** There is a very small number of outliers above the upper whisker, representing a few homes with significantly larger living areas.

- **Interpretation:** The slight right skew, along with a centrally positioned and relatively small box, suggests that most homes have a living space close to the median. The similar lengths of the upper and lower whiskers indicate some variability on both ends, with a few larger properties stretching the distribution slightly higher. This reflects a market where mid-sized homes are prevalent, with some variation toward both smaller and larger homes.



```python
# Boxplot for 'sqft_lot'
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['sqft_lot'])
plt.title('Boxplot of Sqft Lot')
plt.ylabel('Sqft Lot')
plt.show()
```

### Sqft Lot

- **Median:** The median square footage of lot size is positioned closer to the upper quartile (Q3), indicating that a large proportion of properties have lot sizes below this value, with fewer larger lots pulling the median upwards. The median is around 7,000 square feet.

- **Box Size:** The box is reasonably compact, showing that many properties have lot sizes close to the median, with limited variability in this central range.

- **Skewness:** The distribution is slightly right-skewed, as indicated by the longer right whisker. This means that while the majority of lot sizes are clustered around the median, there are some properties with much larger lots that extend the distribution towards higher values. Despite the median being closer to Q3, the longer right whisker suggests a stretch in larger lot sizes.

- **Outliers:** There is a higher variance in outliers above Q4, with some properties having significantly larger lot sizes, contributing to the overall right skew.

- **Interpretation:** The slight right skew, along with the compact box and the longer right whisker, suggests that most properties have smaller to moderately sized lots around 7,000 square feet. However, there are a number of larger lots that extend the distribution to the right, particularly in the upper quartile, where some significantly larger properties are observed.



```python
# Boxplot for 'sqft_above'
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['sqft_above'])
plt.title('Boxplot of Sqft Above')
plt.ylabel('Sqft Above')
plt.show()
```

### Sqft Above

- **Median:** The median square footage above ground is around 1,400 square feet, indicating that half of the properties have an above-ground living space smaller than this value, while the other half are larger.

- **Box Size:** The box is fairly small, indicating that a large number of properties have above-ground square footage close to the median, with limited variability in this central range.

- **Skewness:** The distribution is right-skewed, as shown by the relatively short distance between Q1 and the median, and a longer right whisker. This suggests that while most homes are clustered around the lower to mid-range sizes, there are some properties with larger above-ground living spaces that extend the distribution towards higher values.

- **Whisker Length:** The right whisker is fairly long, though not excessively so, indicating that there is a noticeable stretch of larger homes.

- **Outliers:** There are some outliers above the top of the right whisker, representing a few properties with significantly larger above-ground square footage.

- **Interpretation:** The right-skewed distribution, compact box, and relatively long right whisker indicate that most homes have above-ground living spaces around 1,400 square feet, with a few properties extending the range towards larger sizes. The presence of outliers suggests that there are some homes with much larger above-ground square footage, though they are not common.
