```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/workspaces/Practise-Code/Kaggle Project/Data/Cleaned Data/df_cleaned.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['actual_bathrooms'] = df['bathrooms'].round(0).astype(int)

df_price_by_bathrooms = df.groupby('actual_bathrooms')['price'].mean().reset_index() 

plt.figure(figsize=(12, 6))

sns.barplot(x='actual_bathrooms', y='price', data=df_price_by_bathrooms, palette='viridis', hue='actual_bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Price')
plt.title('Mean Price by number of Bathrooms')
```

**Bathrooms Effect on Mean Price**

- The analysis shows that properties with more bathrooms tend to have higher mean prices.
- The price increase between each additional bathroom is relatively even, indicating a steady rise in mean value with more bathrooms.
- This suggests that adding a bathroom directly correlates with an increase in the mean property value, showing that buyers are willing to pay more for homes with more bathrooms.
- It's important to note that in the original data, half bathrooms (toilette without a shower) were counted as 0.5 bathrooms. However, they have now been rounded up to count as full bathrooms in this analysis.




```python
df_price_by_sqft = df.groupby('sqft_living')['price'].mean().reset_index() # Grouping by qft_living and taking the mean of price

plt.figure(figsize=(12, 6))

sns.scatterplot(x='sqft_living', y='price', data=df_price_by_sqft, palette='viridis', hue='sqft_living')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Price by Square Footage of Living Space')
plt.grid(True)
```

**Price by Square Footage**

- The graph shows a general upward diagonal trend, indicating a positive correlation between square footage and price.
- This suggests that, in general, as square footage increases, property prices tend to rise.
- However, the correlation is not perfectly precise due to the presence of many outliers.
- These outliers may be explained by factors such as:
  - Outdoor area space
  - Property grade
  - Area
  - Year of Proporty Construction



```python
df_price_by_lat = df.groupby('lat')['price'].mean().reset_index() # Grouping by qft_living and taking the mean of price

plt.figure(figsize=(12, 6))

sns.scatterplot(x='lat', y='price', data=df_price_by_lat, palette='coolwarm', hue='lat')
plt.xlabel('Latitude')
plt.ylabel('Price')
plt.title('Price by Latidual Location of Property')
plt.grid(True)
```

**Price by Latitude**

- The data suggests that lower latitude is generally associated with lower home prices.
- There is a noticeable price spike around latitude 47.64, indicating that properties in this region may be more desirable or valuable.
- After this spike, prices tend to taper off as latitude increases, though properties at higher latitudes still tend to be priced higher than those at lower latitudes.
- This trend signifies that certain areas around latitude 47.64 may be considered more desirable, driving up property prices.
