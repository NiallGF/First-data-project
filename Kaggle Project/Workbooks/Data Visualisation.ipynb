```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/niallflaherty/Documents/GitHub/Practise-Code/Kaggle Project/Data/supermarket_sales_clean.csv')
```

```python
df_sales_by_gender = df.groupby('Gender')['Total'].sum()
df_sales_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title('Sales By Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
```

```python
df_sales_by_city = df.groupby('City')['Total'].sum()
df_sales_by_city.plot(kind='bar', color=['blue', 'pink', 'green'])
plt.title('Sales By City')
plt.xlabel('City')
plt.ylabel('Total Sales')
```

```python
df_Naypyitaw = df[df['City'] == 'Naypyitaw']
df_Naypyitaw_sales_by_gender = df_Naypyitaw['Gender'].value_counts()

df_Naypyitaw_sales_by_gender.plot(kind='bar', color=['pink', 'blue'])
plt.title('Gender Sales in Naypyitaw')
plt.xlabel('Gender')
plt.ylabel('Sales Counts')
```

```python
df_Mandalay = df[df['City'] == 'Mandalay']
df_Mandalay_sales_by_gender = df_Mandalay['Gender'].value_counts()

df_Mandalay_sales_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Sales in Mandalay')
plt.xlabel('Gender')
plt.ylabel('Sales Counts')
```

```python
df_Yangon = df[df['City'] == 'Yangon']
df_Yangon_sales_by_gender = df_Yangon['Gender'].value_counts()

df_Yangon_sales_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Sales in Yangon')
plt.xlabel('Gender')
plt.ylabel('Sales Counts')
```
